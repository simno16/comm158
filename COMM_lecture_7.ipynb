# Iteration and functions work together to avoid repeating yourself.

1. Iteration provides input to a function, one at a time.
2. A function contains the block of code that is executed repetitively, producing different outputs for different inputs.

# Functions are easy to maintain. For any change, you only need to implement it once. Then all the outcomes from calling this function will be changed.

# calculate all final scores using a `for-loop` without a function.

1. Code can become messy quickly, even for simple tasks.
2. Nested for loops are hard to read and maintain.



````python
allFinalScores = []
for oneStudentScores in courseScores:
  weightedScores = []
  for i in range(len(oneStudentScores)):
    score = oneStudentScores[i]
    w = weights[i]
    weightedScore = score * w
    weightedScores.append(weightedScore)
  finalScore = sum(weightedScores)
  allFinalScores.append(finalScore)
  print(finalScore)

````

````python
def getFinalScore(oneStudentScores):
  weightedScores = []
  for i in range(len(oneStudentScores)):
    score = oneStudentScores[i]
    w = weights[i]
    weightedScore = score * w
    weightedScores.append(weightedScore)
  finalScore = sum(weightedScores)
  return finalScore
````

````python
allFinalScores = []
for oneStudentScores in courseScores:
  finalScore = getFinalScore(oneStudentScores)
  allFinalScores.append(finalScore)
  print(finalScore)
````
