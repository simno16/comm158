4. A boolean variable stores the result of a statement as True or False. For example, if `x = 5` and `y = 8`, the statement `x < y` evaluates to True, while x > y evaluates to False. In Python, you can write `isSmaller = x < y`, where isSmaller is a boolean variable holding the result of whether `x < y` is True or False.
**TODO: Declare two integer variables and use comparison operators (<, >, ==, !=) to test relationships between them, such as whether one is greater, equal, or not equal to the other. Store each comparison result in a boolean variable, print the outcomes, and add comments explaining each test.**

```python
#assign integer variables
x=5
y=8

isGreater=x>y #test whether x is greater than y
isSmaller=x<y #test whether x is less than y
isEqual=x==y #test whether x is equal to y
isNotEqual=x!=y #test whether x is not equal to y

#print results
print(isGreater)
print(isSmaller)
print(isEqual)
print(isNotEqual)
```
