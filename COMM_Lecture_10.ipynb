```python
names = ["Alice", "Bob", "Charlie", "David", "Eve"]
allTests = [[90, 80, 100], [75, 98, 82], [88, 89, 92], [98, 85, 95], [99, 95, 93]]
```

```python
nameAlltests=dict(zip(names, allTests))
nameAlltests
```

```python
def getAverage(testScores):
  return sum(testScores)/len(testScores)
```

```python
highestAverage=max(allTests, key=getAverage)
highestAverage
```

```python
import numpy as np
```

```python
def getKeyAverage(key):
  value=nameAlltests[key]
  return np.mean(value)
```

```python
topStudent=max(nameAlltests, key=getKeyAverage)
topStudent
```

```python
topStudent=max(nameAlltests, key=lambda name:np.mean(nameAlltests[name]))
topStudent
```

```python
topStudentFirstQuiz=max(nameAlltests, key=lambda name:nameAlltests[name][0])
topStudentFirstQuiz
```

```python
topStudentFirstQuiz=max(nameAlltests, key=lambda name:nameAlltests[name][1])
topStudentFirstQuiz
```
